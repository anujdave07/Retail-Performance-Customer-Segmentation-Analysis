```python
df["order_date"] = pd.to_datetime(df["order_date"], errors="coerce")
df["order_date"].isna().sum()
df.dtypes
```

```python
df["revenue"] = df["price"] * df["quantity"]
df[["price", "quantity", "revenue"]].head()

```

```python
(price_issues := (df["price"] <= 0).sum())
(quantity_issues := (df["quantity"] <= 0).sum())
(revenue_issues := (df["revenue"] <= 0).sum())

(review_low := (df["review_score"] < 0).sum())
(review_high := (df["review_score"] > 5).sum())

(age_low := (df["age"] < 10).sum())
(age_high := (df["age"] > 100).sum())

price_issues, quantity_issues, revenue_issues, review_low, review_high, age_low, age_high
```

```python
total_revenue = df["revenue"].sum()
total_orders = len(df)   # Since i don't have order_id in raw file, each row = one order
total_customers = df["customer_id"].nunique()

aov = total_revenue / total_orders

profit = total_revenue * 0.30   # 30% margin as per assumption
profit_margin = profit / total_revenue

total_revenue, total_orders, total_customers, aov, profit, profit_margin
```

```python
# Create monthly period column
df["order_month"] = df["order_date"].dt.to_period("M").dt.to_timestamp()

# Aggregate by month
monthly_perf = df.groupby("order_month").agg(
    monthly_revenue=("revenue", "sum"),
    monthly_orders=("customer_id", "count")
).reset_index()

# Month-over-Month Growth
monthly_perf["mom_growth_pct"] = monthly_perf["monthly_revenue"].pct_change() * 100

monthly_perf
```

```python
city_perf = df.groupby("city").agg(
    revenue=("revenue", "sum"),
    orders=("customer_id", "count"),
    customers=("customer_id", "nunique")
).reset_index()

city_perf["aov"] = city_perf["revenue"] / city_perf["orders"]

city_perf = city_perf.sort_values(by="revenue", ascending=False)
city_perf
```

Category that makes more money 

```python
category_perf = df.groupby("category_name").agg(
    revenue=("revenue", "sum"),
    orders=("customer_id", "count"),
    customers=("customer_id", "nunique")
).reset_index()

category_perf["aov"] = category_perf["revenue"] / category_perf["orders"]

category_perf = category_perf.sort_values(by="revenue", ascending=False)
category_perf
```

```python
gender_perf = df.groupby("gender").agg(
    revenue=("revenue", "sum"),
    orders=("customer_id", "count"),
    customers=("customer_id", "nunique")
).reset_index()

gender_perf["aov"] = gender_perf["revenue"] / gender_perf["orders"]

gender_perf = gender_perf.sort_values(by="revenue", ascending=False)
gender_perf
```

```python
#finding the age group which spends more

bins = [0, 25, 35, 45, 60, 120]
labels = ["18–25", "26–35", "36–45", "46–60", "60+"]

df["age_bucket"] = pd.cut(df["age"], bins=bins, labels=labels, right=True)

age_perf = df.groupby("age_bucket").agg(
    revenue=("revenue", "sum"),
    orders=("customer_id", "count"),
    customers=("customer_id", "nunique")
).reset_index()

age_perf["aov"] = age_perf["revenue"] / age_perf["orders"]

age_perf
df.to_csv("sales_clean.csv", index=False)
```
